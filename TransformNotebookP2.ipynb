```
from pyspark.sql.types import *
```

```
CarParkBayRestrictions_df = spark.read.json("dbfs:/mnt/interim/CarParkBayRestrictions")
CarParkBayRestrictions_df.createOrReplaceTempView("CarParkBayRestrictions_df")
```

```
%sql

select * from CarParkBayRestrictions_df
```

```
%sql

select distinct typedesc1 from CarParkBayRestrictions_df
```

```
%sql

select bayid,
case
when fromday1 = '0' then "Sunday"
when fromday1 = '1' then "Monday"
when fromday1 = '2' then "Tuesday"
when fromday1 = '3' then "Wednesday"
when fromday1 = '4' then "Thursday"
when fromday1 = '5' then "Friday"
when fromday1 = '6' then "Saturday"
end StartingDay, starttime1 as StartTime,
case 
when today1 = '0' then "Sunday"
when today1 = '1' then "Monday"
when today1 = '2' then "Tuesday"
when today1 = '3' then "Wednesday"
when today1 = '4' then "Thursday"
when today1 = '5' then "Friday"
when today1 = '6' then "Saturday"
end EndingDay, 
endtime1 as EndingTime, typedesc1 as ParkingType from CarParkBayRestrictions_df where typedesc1 like "%Disabled Only"
```
